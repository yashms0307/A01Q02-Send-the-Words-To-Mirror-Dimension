Question 2

Write a Python program that accepts a word from the user and reverse it.

```python
word_to_reverse = input("Input a word to reverse: ")

for character in range(len(word_to_reverse) - 1, -1, -1):
    print(word_to_reverse[character], end="")
print(': is the reversed word' , "\n",)
```
